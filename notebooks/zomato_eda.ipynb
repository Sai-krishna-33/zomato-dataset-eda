# **Import libraries**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Load the dataset**

```python
df = pd.read_csv("zomato.csv", encoding="latin-1")
df.head()
```

# **View basic information**

```python
df.info()
```

```python
print("Rows and Columns:", df.shape)
```

# **Checking missing values**

```python
df.isnull().sum()
```

# **Checking Duplicates**

```python
print("Duplicate Rows:", df.duplicated().sum())
```

# **Data filtering**

```python
df.describe()
```

```python
df.select_dtypes(include=np.number).columns
```

# **UNIVARIATE ANALYSIS**

# **Boxplot**(for numerical)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Restaurant ID'], kde=True)
plt.title("Distribution of Restaurant ID")
plt.xlabel("Restaurant ID")
plt.ylabel("Count")
plt.show()
```

# **BIVARIATE ANALYSIS**

# **Scatterplot**(Numerical vs Numerical)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Price range', y='Votes', data=df)
plt.title("Price range vs Votes")
plt.show()
```

# **Boxplot**(Categorical vs Numerical)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Rating text', y='Votes', data=df)
plt.title("Rating text vs Votes")
plt.show()
```

# **Correlation**(Heat map)

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

# **Outliers detection**

```python
plt.figure(figsize=(8,5))
sns.boxplot(df['Aggregate rating'])
plt.title("Outliers in Aggregate rating")
plt.show()
```
